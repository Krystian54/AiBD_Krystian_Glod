Zaimportowanie potrzebnych bibliotek:

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

Definicja funkcji:

```python
def funkcja(x):
    return x*x+5
```

Definicja argumentów i wyliczenie wartości:

```python
x1 = np.linspace(-1,1)
x2 = np.linspace(-6,6)
x3 = np.linspace(0,5)
y1 = funkcja(x1)
y2 = funkcja(x2)
y3 = funkcja(x3)
```

Wyrysowanie funkcji dla zadanych argumentów:

```python
plt.plot(x1, y1,'r',label='wartości funkcji')
plt.legend()
plt.title('f(x) = x^2 + 5  dla  x > -1 oraz x < 1')
plt.xlabel('x - argumenty')
plt.ylabel('y - wartości funkcji')
plt.show()
plt.plot(x2, y2,'g',label='wartości funkcji')
plt.legend()
plt.title('f(x) = x^2 + 5  dla  x > -6 oraz x < 6')
plt.xlabel('x - argumenty')
plt.ylabel('y - wartości funkcji')
plt.show()
plt.plot(x3, y3,'b',label='wartości funkcji')
plt.legend()
plt.title('f(x) = x^2 + 5  dla  x > 0 oraz x < 5')
plt.xlabel('x - argumenty')
plt.ylabel('y - wartości funkcji')
plt.show()
```
