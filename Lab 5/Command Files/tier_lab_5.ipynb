```python
import pandas as pd

#import i wypisanie danych przez pandas
earthquake_data = pd.read_csv("C:/Users/Krystian/Documents/GitHub/AiBD_Krystian_Glod/Lab 5/Original Data/earthquake_data.csv")
earthquake_data
```

```python
#zmiana nazw kolumn
earthquake_data_renamed_columns = earthquake_data.rename(columns=
                {'In general, how worried are you about earthquakes?': "worried about earthquakes",
                'How worried are you about the Big One, a massive, catastrophic earthquake?': "worried about the Big One",
                'Do you think the "Big One" will occur in your lifetime?': "Big One during lifetime probability", 
                'Have you ever experienced an earthquake?': "experienced of earthquake",
                'Have you or anyone in your household taken any precautions for an earthquake (packed an earthquake survival kit, prepared an evacuation plan, etc.)?': "earthquake prevention",
                'How familiar are you with the San Andreas Fault line?': "knowledge about San Andreas Fault line",
                'How familiar are you with the Yellowstone Supervolcano?': "knowledge about Yellowstone Supervolcano",
                'Age': "age",
                'What is your gender?': "gender",
                'How much total combined money did all members of your HOUSEHOLD earn last year?': "household earnings",
                'US Region': "region"})
earthquake_data_renamed_columns
```

```python
#zapisanie wynikowej tabeli do pliku
earthquake_data_renamed_columns.to_csv(path_or_buf="C:/Users/Krystian/Documents/GitHub/AiBD_Krystian_Glod/Lab 5/Analysis Data/earthquake_data_renamed_columns.csv", index=False)
```

```python
#wyselekcjonowanie trzech kolumn wymaganych w poleceniu
earthquake_data_age_gender_big_one =  earthquake_data_renamed_columns[['age', 'gender', 'Big One during lifetime probability']].sort_values('age')
earthquake_data_age_gender_big_one
```

```python
#zapisanie wynikowej tabeli do pliku
earthquake_data_age_gender_big_one.to_csv(path_or_buf="C:/Users/Krystian/Documents/GitHub/AiBD_Krystian_Glod/Lab 5/Analysis Data/earthquake_data_selected.csv", index=False)
```
